```python
df = pd.read_csv('/content/Attrition.csv')
df.head(2)
```

```python
# Separate categorical columns and Numerical columns . When you have separate
# categorical data then you have to replace its sub-categories to numbers
# i.e. has_covid[yes,no] ---> has_covid[0,1]
```

```python
df.dtypes
```

```python
numerical_df = df.select_dtypes(include = 'int64')
numerical_df.sample()
```

```python
categorical_df = df.select_dtypes(include = 'object')
categorical_df.sample()
```

```python
categorical_df.columns
```

```python
print(categorical_df['Attrition'].unique())
print(categorical_df['BusinessTravel'].unique())
print(categorical_df['Department'].unique())
print(categorical_df['EducationField'].unique())
print(categorical_df['Gender'].unique())
print(categorical_df['JobRole'].unique())
print(categorical_df['MaritalStatus'].unique())
print(categorical_df['Over18'].unique())
print(categorical_df['OverTime'].unique())
```

```python
categorical_df['Attrition'] = categorical_df['Attrition'].map({'Yes':1,'No':0})
categorical_df['BusinessTravel'] = categorical_df['BusinessTravel'].map({'Travel_Rarely':1,'Travel_Frequently':0,'Non-Travel':2})
categorical_df['Department'] = categorical_df['Department'].map({'Sales':0,'Research & Development':1,'Human Resources':2})
categorical_df['EducationField'] = categorical_df['EducationField'].map({'Life Sciences':0,'Medical':1,'Marketing':2,'Technical Degree':3,'Human Resources':4, 'Other':5})
categorical_df['Gender'] = categorical_df['Gender'].map({'Female':0,'Male':1})
categorical_df['JobRole'] = categorical_df['JobRole'].map({'Sales Executive':0, 'Research Scientist':1, 'Laboratory Technician':2,
                                                           'Manufacturing Director':3, 'Healthcare Representative':4, 'Manager':5,
                                                           'Sales Representative':6, 'Research Director':7, 'Human Resources':8})
categorical_df['MaritalStatus'] = categorical_df['MaritalStatus'].map({'Single':0,'Married':1,'Divorced':2})
categorical_df['Over18'] = categorical_df['Over18'].map({'Y':1,'N':0})
categorical_df['OverTime'] = categorical_df['OverTime'].map({'Yes':1,'No':0})
categorical_df.info()
```

```python
# 2. Check if there are missing values are or not ? If there are then handle missing values .
df.isnull().sum()
```

there are no missing values

```python
# You have to rename 3 column names of Numerical columns .
numerical_df = numerical_df.rename(columns={'Age':'updated_age',
                             'Education':'updated_education',
                             'EmployeeCount':'updated_employee_count'})
numerical_df.info()
```

```python
# You have to filter 15-25 rows with all columns of Categorical  columns .
categorical_df.iloc[15:25]
```

```python
# You have to check statically view of Nummerical columns .
numerical_df.describe()
```

```python
# You have to export categorical data into csv format and Nummerical data
# into excel format .
categorical_df.to_csv('categoricals.csv')
numerical_df.to_excel('numericals.xlsx')
```

```python
# In your Numerical data , you have to arrange data in ascending order using
# a single column .
numerical_df = numerical_df.sort_values(by='updated_age')
```
