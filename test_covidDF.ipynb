```python
import pandas as pd
import numpy as np
```

```python
# 1. Separate categorical columns and Numerical columns .
# When you have separate categorical data then you have to replace its sub-categories
# to numbers i.e. has_covid[yes,no] ---> has_covid[0,1] .

df = pd.read_csv("covid_toy.csv")
df.head()
```

```python
df.info()
```

```python
categorical_df = df.select_dtypes(include=['object'])
numerical_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
categorical_df.info()
```

```python
numerical_df.info()
```

```python
print(categorical_df['gender'].unique())
print(categorical_df['cough'].unique())
print(categorical_df['city'].unique())
print(categorical_df['has_covid'].unique())
```

```python
categorical_df['gender'] = categorical_df['gender'].map({'Male':0, 'Female':1})
categorical_df['cough'] = categorical_df['cough'].map({'Mild':0, 'Strong':1})
categorical_df['city'] = categorical_df['city'].map({'Kolkata':0, 'Delhi':1, 'Mumbai':0, 'Bangalore':1})
categorical_df['has_covid'] = categorical_df['has_covid'].map({'No':0, 'Yes':1})
```

```python
categorical_df.info()
```

```python
categorical_df.head()
```

```python
# 2. Check if there are missing values are or not ? If there are then handle missing values .

df.isnull().sum()
```

fever have 10 missing values

```python
df['fever'] = df['fever'].fillna(df['fever'].mean())
```

```python
# 3. You have to rename 3 column names of Numerical columns .
numerical_df = numerical_df.rename(columns={'age':'updated_age', 'fever':'updated_fever'})
numerical_df.head()
```

```python
# 4. You have to filter 15-25 rows with all columns of Categorical columns .
categorical_df.iloc[15:25]
```

```python
# 5. You have to check statically view of Nummerical columns .
numerical_df.describe()
```

```python
# 6. You have to export categorical data into csv format and Nummerical data into excel format .
categorical_df.to_csv('categorical_data.csv')
numerical_df.to_excel('numerical_data.xlsx')
```
