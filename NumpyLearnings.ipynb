```python
# Unique()
'''
It will create an array in which it will remove duplicate values and
it will return 3 arrays:
arr ---> Unique elements of array
return_index = True -----> Indexing od unique element
return_counts = True -----> Frequency of each unique element.
'''

a = np.array([1,2,3,1,1,1,1,4,5,4,4,4,4,6,7])
a
```

```python
b = np.unique(a, return_index =True, return_counts = True)
b
```

```python
# linspace()
'''
It will return an array in which elements
will be same gap in a given range.
syntax:
np.linspace(min, max, stepup_totalnumbers)
'''
a = np.linspace(1, 10, 5)
print(a)
```
